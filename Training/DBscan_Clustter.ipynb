```python
import pandas as pd
```

```python
#small dataset for demo purposes 
data={ 'CustomerID':[1,2,3,4,5,6,7,8],
       'AnnualIncome':[15,16,17,28,30,45,55,60],
       'SpendingScore':[39,81,6,77,40,50,42,30]
     }
```

```python
df = pd.DataFrame(data)
df
```

```python
#import the lib
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#select features for the clustering
X = df[['AnnualIncome','SpendingScore']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
dbscan = DBSCAN(eps=0.8, min_samples=2) #play with eps,
df['Cluster'] = dbscan.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='AnnualIncome', y='SpendingScore', hue='Cluster', palette='Set1',s=500)
plt.title('Customer Segments with DBSCAN')
plt.grid(True)
plt.show()
```
